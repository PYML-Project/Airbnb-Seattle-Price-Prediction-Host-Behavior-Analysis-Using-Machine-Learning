```python
def simplify_location(loc):
    if pd.isnull(loc):
        return "Unknown"
    elif "Seattle" in loc:
        return "Seattle"
    elif "Washington" in loc:
        return "Other Washington"
    elif "United States" in loc:
        return "Other USA"
    else:
        return "International"

listings['host_location_group'] = listings['host_location'].apply(simplify_location)
```

```python

listings = listings[listings['host_location_group'] != 'Unknown']
listings['host_location_group'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(categories=[['Other USA', 'Other Washington','International','Seattle']])
listings['host_location'] = encoder.fit_transform(listings[['host_location_group']])
listings.drop(columns=['host_location_group'], inplace=True)
```

```python
# Handle NaNs if any 
listings['host_response_time'] = listings['host_response_time'].fillna('Unresponsive') 
print(listings['host_response_time'].value_counts())
listings['host_response_time'].isnull().sum() 
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Define the order from slowest to fastest
response_order = [
    'Unresponsive',
    'a few days or more',
    'within a day',
    'within a few hours',
    'within an hour'
]

# Initialize encoder
encoder = OrdinalEncoder(categories=[response_order])

# Apply encoding
listings['host_response_time'] = encoder.fit_transform(
    listings[['host_response_time']]
)

```

```python
# Handle NaNs: fill with 0 
listings['host_response_rate'] = listings['host_response_rate'].fillna('0%')
listings['host_acceptance_rate'] = listings['host_acceptance_rate'].fillna('0%')

# Remove '%' and convert to float
listings['host_response_rate'] = listings['host_response_rate'].str.replace('%', '').astype(float)
listings['host_acceptance_rate'] = listings['host_acceptance_rate'].str.replace('%', '').astype(float)
```

```python
# List of binary columns
binary_cols = ['host_is_superhost', 'has_availability', 'host_identity_verified','instant_bookable']

for col in binary_cols:
    print(f"\nUnique values in {col}:")
    print(listings[col].unique())
    print(listings[col].isnull().sum())

for col in binary_cols:
    listings[col] = listings[col].map({'t': 1, 'f': 0}).astype(int)

```

```python
# 1. Strip the brackets and split the string by comma
# This approach is good, but let's make it a little more robust to handle edge cases
listings['host_verifications'] = listings['host_verifications'].str.strip('[]').str.split(', ').fillna('')

# 2. Explode the lists into new rows
exploded_listings = listings.explode('host_verifications')

# 3. Clean up and handle potential nulls
exploded_listings['host_verifications'] = exploded_listings['host_verifications'].str.strip("''").str.strip()
exploded_listings = exploded_listings[exploded_listings['host_verifications'] != '']
exploded_listings['host_verifications'] = exploded_listings['host_verifications'].replace('nan', '')

# 4. One-hot encode the exploded column
verifications_encoded = pd.get_dummies(exploded_listings['host_verifications'], prefix='verification', dtype=int)

# 5. Group by the original index and sum to get the final one-hot encoded columns
verifications_final = verifications_encoded.groupby(verifications_encoded.index).sum()

# 6. Concatenate the new columns with the original dataframe and drop the old column
# The concatenation must happen BEFORE you drop the original column to avoid errors
listings = pd.concat([listings, verifications_final], axis=1)
listings = listings.drop('host_verifications', axis=1)
```

```python
# List of price-related columns to clean
price_cols = ['price', 'weekly_price', 'monthly_price', 'cleaning_fee', 'extra_people']

# Loop through each column to clean and convert
for col in price_cols:
    # Remove '$' and ',' and convert to float
    listings[col] = listings[col].str.replace('[$,]', '', regex=True).astype(float)
```

```python
# Impute missing values with the median of each column
# This is a better approach than using the mean, as it's less sensitive to outliers.
for col in ['weekly_price', 'monthly_price', 'cleaning_fee']:
    listings[col]=listings[col].fillna(listings[col].median())
```

```python
listings['amenities'] = listings['amenities'].fillna('{}')

# Define a list of key amenities
key_amenities = [
    'TV', 'Cable TV', 'Internet', 'Wireless Internet', 'Air Conditioning',
    'Kitchen', 'Heating', 'Washer', 'Dryer', 'Family/Kid Friendly',
    'Free Parking on Premises', 'Pets Allowed', 'Hot Tub', 'Indoor Fireplace','Breakfast', 'Essentials', 'Shampoo',
    'Smoke Detector', 'Carbon Monoxide Detector', 'First Aid Kit', 'Fire Extinguisher',
    'Family/Kid Friendly', 'Pets Allowed', 'Laptop Friendly Workspace', 'Wheelchair Accessible'
]

# Create a binary column for each key amenity
for amenity in key_amenities:
    # Use str.contains() to check if the amenity string is in the text
    # .astype(int) converts True/False to 1/0
    column_name = f"has_{amenity.lower().replace(' ', '_').replace('/', '_')}"
    listings[column_name] = listings['amenities'].str.contains(amenity, case=False, na=False).astype(int)

# Drop the original 'amenities' column
listings = listings.drop('amenities', axis=1)
```
