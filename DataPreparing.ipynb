```python
def detect_outliers_boxplot(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[col] < lower_bound) | (df[col] > upper_bound)]

for column_group in [price_related_columns, reviews_related_columns, numeric_columns]:
    for col in column_group:
        outliers = detect_outliers_boxplot(df, col)
        if not outliers.empty:
            print(f"Outliers in {col}: {len(outliers)} rows")
            print(f"Min outlier: {outliers[col].min()}  |  Max outlier: {outliers[col].max()}")


for column_group in [price_related_columns, reviews_related_columns, numeric_columns]:
    for col in column_group:
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=df[col])
        plt.title(f'Boxplot of {col}')
        plt.xlabel(col)
        plt.show()
        print(df[col].skew())

    



   
```

```python
List=[]
for column_group in [price_related_columns, reviews_related_columns, numeric_columns]:
    for col in column_group:
        if abs(df[col].skew()) > 1:
            List.append(col)
            print(f"{col} : {df[col].skew()}")
```

```python
#Capping outliers in numeric columns

# List of numeric columns where we want to cap outliers
cap_columns = [
    'price', 'weekly_price', 'monthly_price', 'cleaning_fee',
    'number_of_reviews', 'guests_included', 'extra_people',
    'minimum_nights', 'maximum_nights', 'accommodates', 'bathrooms',
    'bedrooms', 'beds', 'host_response_rate', 'host_acceptance_rate',
    'host_listings_count', 'calculated_host_listings_count'
]

# Capping function
def cap_outliers(df, col, lower_quantile=0.01, upper_quantile=0.99):
    lower_limit = df[col].quantile(lower_quantile)
    upper_limit = df[col].quantile(upper_quantile)
    df[col] = np.where(df[col] < lower_limit, lower_limit, df[col])
    df[col] = np.where(df[col] > upper_limit, upper_limit, df[col])
    return df

# Apply capping to each target column
for col in List:
    if col in df.columns:
        df = cap_outliers(df, col, lower_quantile=0.01, upper_quantile=0.99)


```

```python
moderately_skewed_symmetric=['cleaning_fee', 'review_scores_value', 'cleaning_fee', 'maximum_nights', 'availability_365','price', 'weekly_price', 'monthly_price', 'review_scores_rating', 'accommodates', 'price', 'weekly_price', 'monthly_price', 'extra_people', 'number_of_reviews']
right_skewed = [col for col in moderately_skewed_symmetric if df[col].skew() > 1]
left_skewed = [col for col in moderately_skewed_symmetric if df[col].skew() < -1]
print("Right Skewed Columns:", right_skewed)
print("Left Skewed Columns:", left_skewed)
#Applying Transformation

for col in moderately_skewed_symmetric: 
    print(f"\n{col}:")
    print(f"  Original skew: {df[col].skew():.3f}")
    if col in right_skewed:
        df[col] = np.log1p(df[col])
        print(f"  Logarithmic skew:{df[col].skew():.3f}")
    elif col in left_skewed:
        max_val = df[col].max()
        df[col] = np.log1p(max_val - df[col])
        print(f"  Logarithmic skew:{df[col].skew():.3f}")



```

```python
# Example: columns where log-transform still left high skewness
count_like_columns = ['host_listings_count', 'minimum_nights']

from sklearn.preprocessing import QuantileTransformer
qt = QuantileTransformer(output_distribution='normal')

for col in count_like_columns:
    df[col] = qt.fit_transform(df[[col]])

print(df[['minimum_nights', 'host_listings_count']].skew())
```
